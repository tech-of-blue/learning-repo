# 4.2.4 時系列データ

```python
import pandas as pd
import numpy as np
dates = pd.date_range(start="2025-04-01", 
                      end="2025-04-30")
dates
```

```python
rng = np.random.default_rng(123)
df = pd.DataFrame(rng.integers(1, 31, size=30), 
                  index=dates, 
                  columns=["乱数値"])
df
```

```python
dates = pd.date_range(start="2025-01-01", periods=365)
dates
```

```python
rng = np.random.default_rng(123)
df = pd.DataFrame(
            rng.integers(1, 31, size=365), 
            index=dates, 
            columns=["乱数値"])
df
```

```python
df.groupby(pd.Grouper(freq="ME")).mean()
```

```python
df.loc[:, "乱数値"].resample("ME").mean()
```

```python
pd.date_range(start="2025-01-01", end="2025-12-31", 
              freq="W-SAT")
```

```python
df_year = pd.DataFrame(df.groupby(pd.Grouper(
                  freq="W-SAT")).sum(), columns=["乱数値"])
df_year
```
